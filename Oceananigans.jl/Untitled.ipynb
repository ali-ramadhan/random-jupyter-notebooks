```julia
g = RegularCartesianGrid((512, 1, 512), (20000, 1, 20000); dim=2)
```

```julia
vals = 0.1*exp.(- 100*((g.xC .- g.Lx/2).^2 .+ (g.zC .+ g.Lz/2)'.^2) / (g.Lx^2 + g.Lz^2))
PyPlot.contourf(g.xC, g.zC, vals, vmin=0, vmax=0.1, levels=20); PyPlot.colorbar();
```
