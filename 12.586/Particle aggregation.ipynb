```julia
p = 0.5  # jump probability
N = 1000  # lattice size
Nᵗ = 10000  # number of time steps

m = Array{Int64, 1}(undef, N)
m .= 1

for t in 1:Nᵗ
    for i in 1:N
        if rand() > p  # Particle jump happens.
            m[i%N + 1] += m[i]  # Add mass to next site.
            m[i] = 0  # Current site is now empty.
        end
    end
    for i in 1:N  # Add particles of unit mass to empty sites.
        if m[i] == 0
            m[i] = 1
        end
    end
end
```
