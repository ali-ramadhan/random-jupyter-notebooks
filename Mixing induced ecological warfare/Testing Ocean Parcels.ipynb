```python
import numpy as np
import parcels
import math
from datetime import timedelta
from operator import attrgetter
```

```python
fieldset = parcels.FieldSet.from_parcels("D:\\data\\moving_eddies\\moving_eddies")
```

```python
fieldset.U.show()
```

```python
pset = parcels.ParticleSet.from_list(fieldset=fieldset, pclass=JITParticle, lon=[3.3, 3.3], lat=[46.0, 47.8])
print(pset)
```

```python
pset.show(field=fieldset.U)
```

```python
pset.execute(parcels.AdvectionRK4, runtime=timedelta(days=6), dt=timedelta(minutes=5), output_file=pset.ParticleFile(name="EddyParticles.nc", outputdt=timedelta(hours=1)))
```
