```python
import time

import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
N = 10000

x = np.random.rand(N)
y = np.random.rand(N)
s = np.random.choice([1,2,3], N)

colors = {1: "red", 2: "limegreen", 3: "blue"}
labels = {1: "rock", 2: "paper", 3: "scissors"}

ms = matplotlib.markers.MarkerStyle(marker=".", fillstyle="full")
```

```python
t1 = time.time()

rx, ry = [], []
px, py = [], []
sx, sy = [], []

for i in range(len(s)):
    if s[i] == 1:
        rx.append(x[i])
        ry.append(y[i])
    elif s[i] == 2:
        px.append(x[i])
        py.append(y[i])
    elif s[i] == 3:
        sx.append(x[i])
        sy.append(y[i])
        
plt.scatter(rx, ry, marker=ms, linewidths=0, c=colors[1], edgecolors=colors[1], facecolors=colors[1], s=200)
plt.scatter(px, py, marker=ms, linewidths=0, c=colors[2], edgecolors=colors[2], facecolors=colors[2], s=200)
plt.scatter(sx, sy, marker=ms, linewidths=0, c=colors[3], edgecolors=colors[3], facecolors=colors[3], s=200)

t2 = time.time()
print(t2 - t1)

plt.show()
```

```python
t1 = time.time()

for i in range(len(x)):
    c = colors[s[i]]
    plt.scatter(x[i], y[i], marker=ms, linewidths=0, c=c, edgecolors=c, facecolors=c, s=200)

t2 = time.time()
print(t2 - t1)
    
plt.show()
```

```python
t1 = time.time()

colors = N * [""]

for i in range(len(s)):
    if s[i] == 1:
        colors[i] = "red"
    elif s[i] == 2:
        colors[i] = "blue"
    elif s[i] == 3:
        colors[i] = "limegreen"
        
plt.scatter(x, y, marker=ms, linewidths=0, c=colors, edgecolors=colors, facecolors=colors, s=200)

t2 = time.time()
print(t2 - t1)

plt.show()
```
